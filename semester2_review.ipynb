```python
import requests, random

doc_id = "1VQw8mayLl7C57Z50Wu8ywtmoewCMprd55Mkr9v5KyLU"
url = f"https://docs.google.com/document/d/{doc_id}/export?format=txt"
response = requests.get(url)
answers = []

if response.status_code == 200:
    text = response.text
    lines = text.splitlines()
    sets = [[lines[i], lines[i+1]]for i in range(0, len(lines), 2)]
else:
    print("Failed to fetch document. Status code:", response.status_code)
    exit()

while True:
    input("Press enter to continue\n");
    print("What would you like to do?\n[1] Next question\n[2] Show words that have been used\n[3] Show all answers that have been used\n[4] Show the whole answer list\n[5] Exit")
    go = int(input("\n>"))
    if go == 1 and sets:
        ranNum = random.randint(0, len(sets)-1)
        ranSet = sets[ranNum]
        print(f"\n\n{ranSet[1]}\n")
        input("Press enter to reveal answer: ")
        print(f"Answer: {ranSet[0]}\n\n")
        answers.append(ranSet[0])
        sets.pop(ranNum)
    elif go == 2:
        print()
        for i in sets:
            for k in i:
                print(k)
            print()
        print()
    elif go == 3:
        print("\nAnswers:")
        for i in answers:
            print(i)
        print()
    elif go == 4:
        for i in range(len(sets)):
            print(f"- {sets[i][0]}")
    elif go == 5:
        exit()
    else:
        exit("Enter something valid")
```
