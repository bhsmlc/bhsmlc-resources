## 5. Preprocess our image
This step is necessary to clear inconsistencies between how our model wants the image and how Gradio gives it to us.

```python
import numpy as np, gradio as gr, tensorflow as tf
from tensorflow.keras import models

model = models.load_model("[5]_tensorflow_cnn_model.keras")

def predict(img):
    img = img["composite"]

    # Preprocessing the image for the model
    img = tf.expand_dims(img, -1)          # Add a channel dimension
    img = tf.image.resize(img, (28, 28))   # Resize to 28 by 28 pixels
    img = img[..., :1]                     # Keep only the first channel (grayscale)
    img = np.reshape(img, (1, 28, 28, 1))  # Reshape to match the model's input shape
    img = img / 255.0                      # Normalize pixel values to [0, 1]
    img = 1 - img                          # Invert the image (MNIST digits are white on black)
```

## 7. Give gradio a digit to probability mapping
Gradio will automatically make everything readable and pick the highest probability digit (it'll also show the other ones underneath)

```python
import numpy as np, gradio as gr, tensorflow as tf
from tensorflow.keras import models

model = models.load_model("[5]_tensorflow_cnn_model.keras")

def predict(img):
    img = img["composite"]

    img = tf.expand_dims(img, -1)
    img = tf.image.resize(img, (28, 28))
    img = img[..., :1]
    img = np.reshape(img, (1, 28, 28, 1))
    img = img / 255.0
    img = 1 - img

    prediction = model.predict(img)[0]
    # Return a dictionary mapping each digit (0-9) to its predicted probability
    return {str(i): float(prediction[i]) for i in range(10)}
```

## 8. Create and launch the interface

Create a Gradio Interface and pass configurations to it in its parameters

Advice: don't memorize, you can always come back here or look it up.

```python
import numpy as np, gradio as gr, tensorflow as tf
from tensorflow.keras import models

model = models.load_model("[5]_tensorflow_cnn_model.keras")

def predict(img):
    img = img["composite"]

    img = tf.expand_dims(img, -1)
    img = tf.image.resize(img, (28, 28))
    img = img[..., :1]
    img = np.reshape(img, (1, 28, 28, 1))
    img = img / 255.0
    img = 1 - img

    prediction = model.predict(img)[0]
    return {str(i): float(prediction[i]) for i in range(10)}

# Create and configure a Gradio Interface
interface = gr.Interface(fn=predict, inputs=gr.Sketchpad(image_mode="L", type="numpy"), outputs="label")

# Launch the interface with sharing enabled (Sharing allows you to share your interface)
interface.launch(share=True)
```
