#numpy

```python
import numpy as np
import sys

# Python list
pList = list(range(1000000))
# NumPy array
nArray = np.arange(1000000)

print(pList[0:10])
print(nArray[0:10])
print("Average: " + str(np.mean(nArray)))

mem1 = sys.getsizeof(pList) + sum(sys.getsizeof(i) for i in pList)
mem2 = nArray.nbytes

print(mem1)
print(mem2)
```

#pandas

```python
import pandas as pd

names = np.array(["Alice", "Bob", "Charlie"])
scores = np.array([85, 90, 78])

# Without pandas :(

data = {
    "Name": names,
    "Score": scores
}

print(data)
print() # Just for spacing
print()
print()

# With pandas :D

data = pd.DataFrame({
    "Name": names,
    "Score": scores
})

print(data)
```

#matplotlib

```python
import matplotlib.pyplot as plt

scores = [85,90,78]
names = ["Alice","Bob","Charlie"]

plt.bar(names, scores) #independent, dependent
plt.title("Student Scores")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

#scikit-learn

```python
"""
Dataset:
"Hours_Studied": [2, 4, 6, 8, 10]
"Exam_Score": [46, 61, 73, 92, 95]
"""
```

```python
#import pandas as pd
#import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    "Hours_Studied": [2, 4, 6, 8, 10],
    "Exam_Score": [46, 61, 73, 92, 95]
})

print(data)

model = LinearRegression()
model.fit(data[["Hours_Studied"]], data["Exam_Score"])

predictX = int(input("Enter the number of hours studied\n>"))
output = model.predict([[predictX]])
print("Prediction of exam score: "+str(output))

plt.scatter(data["Hours_Studied"], data["Exam_Score"])
plt.plot(data["Hours_Studied"], model.predict(data[["Hours_Studied"]]))
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```
