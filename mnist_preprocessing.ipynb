# 6. Matplotlib visualizing
Lets see how we can view our data as images

```python
import matplotlib.pyplot as plt

def show_images(dataset, dimensions = (2, 3)):
    fig, axes = plt.subplots(dimensions[0], dimensions[1], figsize=(dimensions[1]*2, dimensions[0]*2))
    axes = axes.flatten()
    for i in range(dimensions[0]*dimensions[1]):
        axes[i].imshow(dataset[i], cmap="gray_r")
        axes[i].set_title(f"{y_train[i]}")
    
    plt.tight_layout()
    plt.plot()
    #plt.imshow(X_train[0], cmap="gray_r")
    #plt.show()
show_images(X_train, (5, 10))
print(X_train.dtype)
```
